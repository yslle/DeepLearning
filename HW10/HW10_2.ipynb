```python
import requests
import pandas as pd
import os
from bs4 import BeautifulSoup

urls = [
    "https://news.naver.com/main/main.nhn?mode=LSD&mid=shm&sid1=100", # 정치
    "https://news.naver.com/main/main.nhn?mode=LSD&mid=shm&sid1=101", # 경제
    "https://news.naver.com/main/main.nhn?mode=LSD&mid=shm&sid1=102", # 사회
    "https://news.naver.com/main/main.nhn?mode=LSD&mid=shm&sid1=103", # 생활/문화
]

#크롤링 자동 방지 해제
headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) \
           AppleWebKit/537.36 (KHTML, like Gecko) Chrome/87.0.4280.88 Safari/537.36"}

# 파일 존재 여부 확인
if os.path.isfile('/content/test_SectionDataset.csv'):
      df = pd.read_csv('/content/test_SectionDataset.csv', names=['title', 'section'], header=0)
else:
      bodyStruc = {'title': [], 'section': []}
      df = pd.DataFrame( bodyStruc, columns=['title', 'section'] )

# 섹션 번호
section_numbers = ['0', '1', '2', '3']
section_num = 0

# URL 목록 순회
for url in urls:
  result = requests.get(url, headers=headers)
  #print(result)

  soup = BeautifulSoup(result.text, 'html.parser')
  #print(soup)

  #news_titles = soup.find_all(class_='cluster_text_headline')
  news_titles = soup.select('.rl_txt')
  #print(news_titles)

  print('총 뉴스 제목 개수: ', len(news_titles))
  # for item in news_titles:
      # print(item.get_text())
      # print(item['href'])

  for item in news_titles:
      texts = item.get_text()
      texts = texts.replace(',', "")
      texts = texts. replace('\'', "")
      texts = texts.replace('\"', "")
      texts = texts.replace('…', " ")
      print(texts, section_numbers[section_num])
      oneData = {'title': texts, 'section': section_numbers[section_num]}
      # df = df.append( oneData, ignore_index=True )
      oneData_df = pd.DataFrame([oneData])  # dict를 DataFrame으로 변환
      df = pd.concat([df, oneData_df], ignore_index=True)
  section_num +=1

df.to_csv('./test_SectionDataset.csv', encoding='UTF-8', index=False, header=True)
```
