```python
from keras.models import Sequential
from keras.layers import Dense
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import StratifiedKFold

import numpy
import pandas as pd
import tensorflow as tf
import os

# seed 값 설정
seed = 0
numpy.random.seed(seed)
tf.random.set_seed(3)

df = pd.read_csv('/content/drive/MyDrive/24-1 인공지능/080228-master/deeplearning/dataset/sonar.csv', header=None)

dataset = df.values
X = dataset[:,0:60].astype(float)
Y_obj = dataset[:,60]

e = LabelEncoder()
e.fit(Y_obj)
Y = e.transform(Y_obj)

# 5개의 파일로 쪼갬
n_fold = 5
skf = StratifiedKFold(n_splits=n_fold, shuffle=True, random_state=seed)

# 빈 accuracy 배열
accuracy = []

# 모델 저장 폴더 설정
MODEL_DIR = './model/'
if not os.path.exists(MODEL_DIR):
  os.mkdir(MODEL_DIR)

index = 0
# 모델의 설정, 컴파일, 실행
for train, test in skf.split(X, Y):
    model = Sequential()
    model.add(Dense(24, input_dim=60, activation='relu'))
    model.add(Dense(10, activation='relu'))
    model.add(Dense(1, activation='sigmoid'))
    model.compile(loss='mean_squared_error',
                  optimizer='adam',
                  metrics=['accuracy'])
    model.fit(X[train], Y[train], epochs=100, batch_size=5)

    # 모델을 컴퓨터에 저장
    model.save('./model/lab7-model-%d.hdf5' % index)
    index = index+1
    k_accuracy = "%.4f" % (model.evaluate(X[test], Y[test])[1])
    accuracy.append(k_accuracy)

# 결과 출력
print("\n %.f fold accuracy:" % n_fold, accuracy)

# 가장 높은 accuracy 가진 모델의 인덱스 출력
best_accuracy = accuracy[0]
best_acc_index = 0

for i in range(index):
  if (accuracy[i] > best_accuracy):
    best_accuracy = accuracy[i]
    best_acc_index = i
print("Max index:", best_acc_index);
```

```python
from keras.models import Sequential, load_model
from keras.layers import Dense
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import StratifiedKFold

import numpy as np
import pandas as pd
import tensorflow as tf

# 데이터 입력
df = pd.read_csv("/content/drive/MyDrive/24-1 인공지능/080228-master/deeplearning/dataset/sonar.csv", header=None)
dataset = df.values
X = dataset[:, 0:60].astype(float)
Y_obj = dataset[:, 60]

# 문자열 변환
e = LabelEncoder()
e.fit(Y_obj)
Y = e.transform(Y_obj)

# 앞부분 5개의 행 추출
test_X = X[:5, :]
test_Y = Y[:5]

# 가장 높은 accuracy를 가진 딥러닝 모델 로드
best_acc_index = 1
model = load_model('./model/lab7-model-%d.hdf5' % best_acc_index)
prediction = model.predict(test_X)
print("Load %s-th model having %.4f accuracy" % (best_acc_index, model.evaluate(X, Y)[1]))

for i in range(len(prediction)):
  print("index: %d, actual: %.2f, expect: %s" % (i, test_Y[i], prediction[i]))
```
