실습과제 1

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# data 읽어오기
df = pd.read_csv('/content/drive/MyDrive/24-1 인공지능/080228-master/deeplearning/dataset/pima-indians-diabetes.csv',
                 names = ["pregnant", "plasma", "pressure", "thickness", "insulin", "BMI", "predigree", "age", "class"])
df.head()
```

```python
# 1, 2번째 열을 배열로 변환하여 추출
df_list = df[['pregnant', 'plasma']].values.tolist()
#print(df_list)
#print(type(df_list))

# 추출한 배열을 2차원 벡터 data1[][2]에 저장
data1 = np.array(df_list)
#print(data1)
#print(type(data1))

# data1에 저장된 5개만 분리하여 2차원 벡터 data2[5][2]에 저장
data1_list = data1.tolist()

data2_list = []
for i in range(5):
  data2_list.append(data1_list[i])
data2 = np.array(data2_list)
print(data2)
print(data2.shape)
print(type(data2))

# data2에 저장된 첫번째 열의 값들을 일차원 배열 x1[]에, 두번째 열의 값들을 y1[]에 저장
x1 = []
y1 = []
for i in range(data2.shape[0]):
  x1.append(data2[i][0])
  y1.append(data2[i][1])

print(x1)
print(y1)
```

```python
# x1, y1을 2차원 그래프로 출력
plt.plot(x1, y1)
plt.plot(x1, y1, "o")
plt.show()
```

```python
# 데이터에서 평균 age 출력
data = df[['age', 'thickness']]
age_mean = data['age'].mean()
print("Average age:", round(age_mean, 2))

# 평균 age 이하, 이상의 thisckness 분포도를 그래프로 출력
df2 = data[data['age'] <= age_mean] # 평균 이하
df3 = data[data['age'] >= age_mean] # 평균 이상

df2.thickness.plot.hist(bins=100, alpha=0.5)
df3.thickness.plot.hist(bins=100, alpha=0.5)
plt.legend(['low_age', 'high_age'], loc="upper right")
plt.xlabel('thickness')
plt.show()
```
